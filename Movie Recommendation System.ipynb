```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data=pd.read_csv('Movie-Recommendation-System\movies.csv')
data
```

```python
data.shape
```

```python
data.info()
```

```python
sfeatures=['genres','keywords', 'tagline', 'cast','director']
sfeatures
```

```python
for i in sfeatures:
    data[i]=data[i].fillna('')
```

```python
data.isnull().sum()
```

```python
cfeatures = data['genres'] + ' ' + data['keywords'] + ' ' + data['tagline'] + ' ' + data['cast'] + ' ' + data['director']
cfeatures
```

```python
tfi=TfidfVectorizer()
```

```python
ftfts=tfi.fit_transform(cfeatures)
print(ftfts)
```

```python
sim=cosine_similarity(ftfts)
```

```python
print(sim)
```

```python
movie_name=input('Enter a your fav movie name:')
```

```python
movies_list=data['title'].tolist()

approp_names=difflib.get_close_matches(movie_name,movies_list)

correct_name=approp_names[0]

correct_index=data[data['title']==correct_name]['index'].values[0]

correct_sim=list(enumerate(sim[correct_index]))

sorted_sim=sorted(correct_sim,key=lambda x:x[1],reverse=True)

i=1
print(f'Recommended movies based on {movie_name}:')
for movie in sorted_sim:
    movie_index=movie[0]
    res=data[data['index']==movie_index]['title'].values[0]
    if i<31:
        print(i,'.',res)
        i=i+1
```
